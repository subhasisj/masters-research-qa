### TAPT
Task Adaptive Pretraining.

Here we will take a Finetuned Model on QA task (Multilingual MiniLMv2), separate the language model for further Out of domain pretraining.

```python
from transformers import AutoModelForQuestionAnswering, BertConfig

model_path = "../../Models/squad-qa-minilmv2-XLMTokeinizer-8/checkpoint-7000"
qa_model = AutoModelForQuestionAnswering.from_pretrained(model_path)

qa_model
```

```python
# Remove the encoder block from QA model
shared_encoder = getattr(qa_model, "bert")
# configuration = BertConfig()
# configuration.vocab_size = qa_model.config.vocab_size

#Load model for Pretraining using Masked Language Modelling
from transformers import AutoModelForMaskedLM
model = AutoModelForMaskedLM.from_pretrained("microsoft/Multilingual-MiniLM-L12-H384")

setattr(model, "bert", shared_encoder)

model
```

```python
model.push_to_hub('MiniLMv2-qa-encoder')
```
